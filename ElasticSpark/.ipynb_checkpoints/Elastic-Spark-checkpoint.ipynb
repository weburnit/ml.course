```python
from datetime import datetime
from elasticsearch import Elasticsearch
```

```python
es = Elasticsearch(['http://es.ops.lzd.co'], port=80)
```

```python
end = datetime.strptime("06/09/2017 16:30", "%d/%m/%Y %H:%M")
start = datetime.strptime("01/09/2017 16:30", "%d/%m/%Y %H:%M")
timerange = {"range": {"@timestamp": {"gte": int(start.timestamp()*1000),"lte": int(end.timestamp()*1000),"format": "epoch_millis"}}}
```

```python
query = 'tag:"kubernetes.var.log.containers.lgms*"'
res = es.search(index='lgms-app*', body={"query": {"bool": {"must":[{"query_string": {"analyze_wildcard": True,"query": query}}, timerange]}}})
```
